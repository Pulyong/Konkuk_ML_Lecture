```
import numpy as np
from hmmlearn import hmm

states = ["Rainy", "Cloudy", "Sunny"]
n_states = len(states)

observations = ["Boots", "Shoes"]
n_observations = len(observations)
############### 모든 것을 train 데이터에서 얻어와야함 ####################
# 시작 확률
start_probability = np.array([0.2, 0.5, 0.3]) # train 데이터에서 얻어와야함

# 전이 확률 MLE로 얻어와야함
# 0.4 = (R and R) / R
transition_probability = np.array([
  [0.4, 0.3, 0.3],
  [0.2, 0.6, 0.2],
  [0.1, 0.1, 0.8]
])

# 관측 확률 state에서 feature가 emission될 확률 (Sunny가 주어졌을 때 shoes가 나올 확률 P(S|sunny), 뒤집히면 generation probability)
emission_probability = np.array([
  [0.8, 0.2],
  [0.5, 0.5],
  [0.1, 0.9]
])

# 모델 만들기
model = hmm.CategoricalHMM(n_components=n_states)
# 여기를 채우세요.
model.startprob_ = start_probability
model.transmat_ = transition_probability
model.emissionprob_ = emission_probability
# 관측 입력
input = [0, 0, 1,1]

# HMM 모델 호출
# 여기를 채우세요.
# input -> (1,4) -> (4,1)
# 2차원으로 바꾼 다음 transpose
hmm_input = np.atleast_2d(input).T
logprob, sequence = model.decode(hmm_input) # -> viterbi algorithm
# 0이되면 안되기 때문에 smoothing
# why log? observation*transition*observation*transition ... -> 곱셈은 오래걸리기 때문에 log scale -> 덧셈으로 변경
# log likelyhood estimation

print("Input :", ", ".join(map(lambda x: observations[x], input))) 
print("Output:", ", ".join(map(lambda x: states[x], sequence)))
print("Prob. :", logprob)
```
